# Clock Slope Approximation
Approximating functions piecewise with lines having power-of-two slopes.

```julia
using Plots;
using Calculus;
```

```julia
t0 = 0.;
tf = 2.;
samples = 1000;
Δt = (tf - t0) / samples;
pts = t0:0.1Δt:tf;
```

```julia
τ(t) = sin(t * 2π * 2 / (tf - t0));
```

```julia
function sapprox(f, t0, Δt, tf)
    values = [];
    t = t0;
    ft = f(t);
    while (t <= tf)
        dft = derivative(f, t);
        push!(values, [ft, dft]);
        ft += Δt * dft;
        t += Δt;
    end
    dft = derivative(f, t);
    push!(values, [ft, dft]);
    x -> (
        param = (x - t0) / Δt;
        dparam = floor(param);
        value = values[Int(dparam) + 1];
        value[1] + value[2] * (param - dparam) * Δt
    )
end
```

```julia
sτ = sapprox(τ, t0, Δt, tf)
```

```julia
smax = 2^20;
function calc_dv(dv)
    p = clamp(round(log2(abs(dv))), 0, smax);
    sign(dv) * 2^p;
end
function dapprox(f, t0, Δt, tf)
    values = [];
    t = t0;
    ft = f(t);
    while (t <= tf)
        dft = calc_dv(derivative(f, t));
        push!(values, [ft, dft]);
        ft += Δt * dft;
        t += Δt;
    end
    dft = calc_dv(derivative(f, t));
    push!(values, [ft, dft]);
    x -> (
        param = (x - t0) / Δt;
        dparam = floor(param);
        value = values[Int(dparam) + 1];
        value[1] + value[2] * (param - dparam) * Δt
    )
end
```

```julia
dτ = dapprox(τ, t0, Δt, tf)
```

```julia
plot(pts, [τ, sτ, dτ], xlims = (t0, tf))
```
